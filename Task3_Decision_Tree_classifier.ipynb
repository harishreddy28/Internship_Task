**Step:1 Import the libraies**

```
# Load libraries
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
```

**Step 2.Import the data**

```
# Load data
iris = datasets.load_iris()
# Forming the iris dataframe
features = iris.data
target = iris.target
```

**Step 3.Training a Decision Tree Classifier using Gini and entropy**

Introduction on Decision tree

In a decision tree, every decision rule occurs at a decision node, with the rule
creating branches leading to new nodes. A branch without a decision rule at the end is called a leaf.


```
# Create decision tree classifier object
decisiontree = DecisionTreeClassifier(random_state=0)
```

```
# Train model
model = decisiontree.fit(features, target)
```

Decision tree attempt to find a decision rule that produces the greatest decrease in impurity at a node. While there are a number of measurements of impurity, by default DecisionTreeClassifier uses Gini impurity

In scikit-learn, DecisionTreeClassifier operates like other learning methods; after
the model is trained using fit we can use the model to predict the class of an observation

```
# Make new observation
observation = [[ 5, 4, 3, 2]]
```

```
# Predict observation's class
model.predict(observation)
```

We can also see the predicted class probabilities of the observation

```
# View predicted class probabilities for the three classes
model.predict_proba(observation)
```

Now we want to use a different impurity measurement we can use the criterion
parameter

```
# Create decision tree classifier object using entropy
decisiontree_entropy = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```
# Train model
model_entropy = decisiontree_entropy.fit(features, target)
```

```
# Predict observation's class
model_entropy.predict(observation)
```

```
# View predicted class probabilities for the three classes
model_entropy.predict_proba(observation)
```

**Step 4.Visualizing a Decision Tree Model**

```
# Load libraries
import pydotplus
from sklearn.tree import DecisionTreeClassifier
from IPython.display import Image
from sklearn import tree
```

```
# Create DOT data
dot_data = tree.export_graphviz(decisiontree,out_file=None,feature_names=iris.feature_names,class_names=iris.target_names,filled=True, rounded=True,special_characters=True)
```

```
# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)
```

```
# Show graph
Image(graph.create_png())
```

One advantages of decision tree classifiers is that we can visualize the entire
trained model—making decision trees one of the most interpretable models in
machine learning. In above solution, I exported  trained model in DOT format (a
graph description language) and then used that to draw the graph

If we look at the root node, we can see the decision rule is that if petal widths are less
than or equal to 0.8, then go to the left branch; if not, go to the right branch. We can
also see the Gini impurity index (0.667), the number of observations (150), the number of observations in each class ([50,50,50]), and the class the observations would
be predicted to be if we stopped at that node (setosa). We can also see that at that
node the learner found that a single decision rule (petal width (cm) <= 0.8) was
able to perfectly identify all of the setosa class observations. Furthermore, with one
more decision rule with the same feature (petal width (cm) <= 1.75) the decision
tree is able to correctly classify 144 of 150 observations.

**This makes petal width a very important feature!**
